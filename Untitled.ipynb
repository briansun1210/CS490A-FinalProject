```python
import numpy as np
import pandas as pd
import torch
import transformers as ppb # pytorch transformers
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
nlp = spacy.load("en_core_sci_sm")
med7 = spacy.load("en_core_med7_lg")
```

```python
def tokenize_doc(text):
    bow = defaultdict(float)
    temp = []
    doc = nlp(text)
    doc = doc.ents
    lowered_tokens = map(lambda t: str(t).lower(), doc)

    for token in lowered_tokens:
        temp.append(token)
#     temp = getNGrams(temp,5)
    temp = " ".join(temp)
    return temp
```

```python
X = [tokenize_doc(x) for x in data_t['transcription']]
X = pd.Series(X)
```

```python
tokenizer = BertTokenizer.from_pretrained('emilyalsentzer/Bio_ClinicalBERT') 
```

```python
# Load pre-trained model (weights)
model = BertModel.from_pretrained('emilyalsentzer/Bio_ClinicalBERT',
                                  output_hidden_states = True, # Whether the model returns all hidden-states.
                                  )
# Put the model in "evaluation" mode, meaning feed-forward operation.
model.eval()
```

```python
tokenized = X.apply((lambda x: tokenizer.convert_tokens_to_ids(tokenizer.tokenize(x))[0:512]))
max_len = 512
tokenized = np.array([i + [0]*(max_len-len(i)) if len(i) < 512 else i for i in tokenized])
label = data_t["medical_specialty"]

print(np.array(tokenized).shape)
tokenized = torch.tensor(tokenized)
print(len(tokenized))
```

```python
print(np.array(tokenized).shape)
label = data_t["medical_specialty"].to_numpy()
```

```python
tokenized = tokenized.to(torch.int64)
with torch.no_grad():
    outputs = model(tokenized)
    hidden_states = outputs[2]
```

```python
token_embeddings = torch.stack(hidden_states, dim=0)

token_embeddings.size()
```

```python
temp = [] 
# `token_vecs` is a tensor with shape [22 x 768]
for i in range (0, len(hidden_states[-2])):
    token_vecs = hidden_states[-2][i]

# Calculate the average of all 22 token vectors.
    sentence_embedding = torch.mean(token_vecs, dim=0)
  # print(sentence_embedding)
    temp.append(sentence_embedding.numpy())
print(temp[0][0])
```

```python
train_features, test_features, train_labels, test_labels = train_test_split(temp, label)
```

```python
lr_clf2 = LogisticRegression()
lr_clf2.fit(train_features, train_labels)
```

```python
lr_clf2.score(test_features, test_labels)
```
